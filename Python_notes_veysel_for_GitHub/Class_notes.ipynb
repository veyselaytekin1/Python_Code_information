## Şimdi de class method tanımlayalım. Dekoratör

```python
# Diyelim ki isimle yaş birlikte verilmiş.   
class Person:
    maas_zammi = 1.8
    p_numbers = 0 
    def __init__(self,name, age):  
        self.name = name 
        self.age = age
        Person.p_numbers += 1  #her bir person ekleyince init bunu aktif hale getiriyor
    
    def show_info(self):                     # Instance metdods
        print(f"name : {self.name}\nage : {self.age}") 
        
# @classmethodu görünce diyor ki, bu class metod        
    @classmethod     # obje üretmeden fonksiyon oluştırma  basina @ isareti demek
    def person_num(cls):   #önceden self yaziyorduk ama class methotlarda cls yapiyoruz
        return cls.p_numbers  #bunu döndürecek
    
    @classmethod   #yani artik nesneye ait degil anlamina geliyor
    def create_with_string(cls,stri_):   # cls PEP8 yazılım kuralı
        name,age= stri_.split(" ")  #fonksiyon adinida stri_'yide biz verdik,cls den sonrakide stri_
        return cls(name,age)
    
person1=Person("Ayşe",80)
person2=Person("Neşe",82)
person3=Person("Ahmet",15)

person4=Person.create_with_string("Aslan 25") #bunu ayirdi ve ayirdiktan sonra 25 i yas olarak atadi
person5=Person("Saghmet",15)

print(Person.person_num())

#class method ne is yapiyor;nesneye ait olmayan bir fonksiyon yazdigimizda kullaniyoruz
```

```python
class Person:
    maas_zammi = 1.8
    p_numbers = 0 
    def __init__(self,name, age):
        self.name = name 
        self.age = age
        Person.p_numbers += 1
    
    def show_info(self):                           # Instance metdods
        print(f"name : {self.name}\nage : {self.age}")
        
# Structor method, yapıcı   
    @classmethod     # obje üretmeden fonksiyon çalıştırma
    def person_number(cls):
        return cls.p_numbers
    
    @classmethod   #bu yukaridaki nesneye ait olmuyor
    def create_with_string(cls,stri_):
        name,age= stri_.split(" ")
        return cls(name,age)
    
    @classmethod
    def age_calc(cls,name,birth_year):
        return cls(name,date.today().year-birth_year)
    
person1=Person("Ayşe",80)
person2=Person("Neşe",82)
person3=Person("Ahmet",15)
person4=Person.create_with_string("Aslan 25")
person5=Person("Saghmet",15)                           
person6=Person.age_calc("Aga",2010)
print(Person.person_number())
```

```python
class Person:
    maas_zammi = 1.8
    p_numbers = 0 
    def __init__(self,name, age):   # Burdaki name ve age class a ait attribute(özellik) değil, instance aittir.
        self.name = name 
        self.age = age
        Person.p_numbers += 1
    
    def show_info(self):                     # Instance metdods
        print(f"name : {self.name}\nage : {self.age}")
        
    @classmethod     # obje üretmeden fonksiyon alıştırma
    def person_number(cls):
        return cls.p_numbers
    
    @classmethod                           # class method
    def create_with_string(cls,stri_):   
        name,age= stri_.split(" ")
        return cls(name,age)
    
    @classmethod
    def age_calc(cls,name,birth_year): 
        return cls(name,date.today().year-birth_year)
    
    @staticmethod # Static method içine değişken almak zorunda değil.
    def birth_year_calc(Person):
        return date.today().year-Person.age
    
person1=Person("Ayşe",80)
person2=Person("Neşe",82)
person3=Person("Ahmet",15)
person4=Person.create_with_string("Aslan 25")
person5=Person("Saghmet",15)                         
person6=Person.age_calc("Aga",2010)
print(Person.person_number())
```

```python
class Person:
    maas_zammi = 1.8
    p_numbers = 0 
    def __init__(self,name, age):   # Burdaki name ve age class a ait attribute(özellik) değil, instance aittir.
        self.name = name 
        self.age = age
        Person.p_numbers += 1
    
    def show_info(self):                     # Instance metdods
        print(f"name : {self.name}\nage : {self.age}")
        
    @classmethod     # obje üretmeden fonksiyon alıştırma
    def person_number(cls):
        return cls.p_numbers
    
    @classmethod                           # class method
    def create_with_string(cls,stri_):   
        name,age= stri_.split(" ")
        return cls(name,age)
    
    @classmethod  
    def age_calc(cls,name,birth_year):
        return cls(name,date.today().year-birth_year)
    
    @staticmethod    
    def birth_year_calc(Person):
        return date.today().year-Person.age
    
    @staticmethod    
    def slm():
        return "Merhaba"
    
person1=Person("Ayşe",80)
person2=Person("Neşe",82) 
person3=Person("Ahmet",15)

person4=Person.create_with_string("Aslan 25")
person5=Person("Saghmet",15)


               
               
person6=Person.age_calc("Aga",2010)
print(Person.person_number())
```

```python
# büyük küçük karşılaştırması  __lt__
class Footballer:
    def __init__(self, name, surname, age):
        self.name = name
        self.surname = surname
        self.age = age
        
    def __eq__(self,other):
        if self.name == other.name and self.surname == other.surname:
            return True
        return False  
    
    def __add__(self,other):
        name = self.name[0] + other.name[0]
        surname = self.surname[0] + other.surname[0]
        age = (self.age + other.age)/2
        return Footballer(name, surname, age)
    
    def __lt__(self,other):    #lees-than _gt_
        if self.age < other.age:
            return True
        return False
    
footballer1=Footballer("Maradona","Süslü",25)
footballer2=Footballer("Zidane","Kop",231) 
```
